```python
""" 

    @Author: Likhitha S
    @Date: 25-10-2024 15:30
    @Last Modified by: Likhitha S
    @Last Modified time: 26-10-2024 15:30
    @Title: Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a url 
    https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv 

"""


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def main():
    """

        Description: 
            This function is used to draw a graph using scatter plot.
        Parameters: 
            url is the input where we are going to get the dataset.
        return:
            It displays the graph with scatterplot.

    """
    
    
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv "
    var= pd.read_csv(url)
    
    print(var)
    sns.scatterplot(x='day', y='total_bill', data=var)
    
    plt.title("Total Bill V/S Day's")
    plt.xlabel("day's")
    plt.ylabel('totalbill')
    
    plt.show()
    
if __name__=="__main__":
    main()
```
