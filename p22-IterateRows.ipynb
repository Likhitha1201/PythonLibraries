```python

""" 
    @Author: Likhitha S
    @Date: 14-10-2024 9:30
    @Last Modified by: Likhitha S
    @Last Modified time: 14-10-2024 9:30
    @Title:Write a Python program to iterate over rows in a DataFrame.  
    
"""

import pandas as pd
import numpy as np
from logging import *

def iterating_data(data):
    """

        Description: 
            This function is used to iterate the rows from dataframes.
        Parameters: 
            data is an complete information of all the exam atire .
        return:
            It prints the iterated data.
            
    """
    
    getLogger().handlers.clear()
    basicConfig(filename="file.log", filemode='w', level=INFO, format=("%(asctime)s: %(levelname)s: %(message)s"))
    info(f"orginal overall data will be: {data} ")
   
    for index, row in data.iterrows():
        info(f"iteration of data will be: {index, row}")
        print(f"iteration of data will be: {index, row}")   
   
def main():
    exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 
    'Laura', 'Kevin', 'Jonas'], 
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19], 
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']} 
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'] 
   
    info(f"original data: {exam_data} and label: {labels}")
    df=pd.DataFrame(exam_data, index=labels)
    iterating_data(df)
    
    
if __name__=="__main__":
    main()
```
