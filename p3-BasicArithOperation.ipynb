```python

""" 
    @Author: Likhitha S
    @Date: 13-10-2024 13:40
    @Last Modified by: Likhitha S
    @Last Modified time: 13-10-2024 13:40
    @Title: Write a Python program to create and display a one-dimensional array-like object containing an array of data using Pandas module.  
           
"""

import pandas as pd
import logging

def add(data1,data2):
    """

        Description: 
            This function is used to add two series.
        Parameters: 
            data1 and data2 are the two different series .
        return:
            It return the output.

    """
    
    logging.debug(f"addition of {data1} and {data2} is:")
    return data1+data2

def sub(data1, data2):
    """

        Description: 
            This function is used to subtract two series.
        Parameters: 
            data1 and data2 are the two different series .
        return:
            It return the output.

    """
    logging.debug(f"subtraction of {data1} and {data2} is:")
    return data1-data2

def div(data1, data2): 
    """

        Description: 
            This function is used to divide two series.
        Parameters: 
            data1 and data2 are the two different series .
        return:
            It return the output.

    """
    logging.debug(f"division of {data1} and {data2} is:")
    return data1/data2

def mult(data1, data2):
    """

        Description: 
            This function is used to multiply two series.logging.debug(f"multiplication of {data1} and {data2} is:")
        Parameters: 
            data1 and data2 are the two different series .
        return:
            It return the output.

    """
    logging.debug(f"multiplication of {data1} and {data2} is:")
    return data1*data2
         
def main():
    logging.basicConfig(filename="file.log", filemode='w'.level=logging.DEBUG, format=("%(asctime)s- %(levelname)s- %(message)s"))
    data1=pd.Series([12,13,14,15,16])
    data2=pd.Series([23,34,45,67,78])
    addition=add(data1,data2)
    
    logging.info(f"after addition the of two series it gives:{addition}")
    print("addition of two series will be:",addition)
    
    division=div(data1, data2)
    logging.info(f"after addition the of two series it it gives:{addition}")
    print("division of two series will be:", division)
    
    subtraction=sub(data1, data2)
    logging.info(f"after subtration the of two series it looks like:{addition}")
    print("subtraction of two series will be:", subtraction)
    multilication=mult(data1, data2)
    print("multiplication of two series will be:", multilication)
    
if __name__=="__main__":
    main()
```
