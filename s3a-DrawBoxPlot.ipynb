```python
""" 

    @Author: Likhitha S
    @Date: 25-10-2024 15:00
    @Last Modified by: Likhitha S
    @Last Modified time: 25-10-2024 15:00
    @Title:  Write a program to draw box plot of life expectancy by continent for a data set given in a 
    url- https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv 
     
"""


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def main():
    """

        Description: 
            This function is used to draw a graph using box plot.
        Parameters: 
            url is the input where we are going to get the dataset.
        return:
            It displays the box graph.

    """
    
    
    url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
    var= pd.read_csv(url)
    
    print(var)
    sns.violinplot(x='continent', y='lifeExp', data=var)
    
    plt.title('Life Expectancy By Continent')
    plt.xlabel('Continent')
    plt.ylabel('Life Expectancy')
    
    plt.show()
    
if __name__=="__main__":
    main()
```
