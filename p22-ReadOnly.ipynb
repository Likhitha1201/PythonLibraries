```python
""" 

    @Author: Likhitha S
    @Date: 10-10-2024 12:50
    @Last Modified by: Likhitha S
    @Last Modified time: 10-10-2024 12:50
    @Title:Write a Python program to make an array immutable (read-only). 

"""

import numpy as np
import logging

def read_array(arr):
    """

        Description: 
            This function is used to create an array with one's and zero's as specified pattern.
        Parameters: 
            arr ia an array passed from main method.
        return:
            It returns an updated array.

    """
    
    logging.basicConfig(filename="common.log", filemode='w', level=logging.DEBUG, format=("%(asctime)s- %(levelname)s- %(message)s"))
    # making array to read only by specifying  below condition
    arr.flags.writeable=False
    return arr
    
def check_immutablity(arr):
    logging.debug("Testing the given array is immutable or not:")
    logging.debug(f"Array: {arr}")
    logging.debug(f"checking the given array is writeable? {arr.flags.writeable} ")
    
    try:
        logging.debug("Now trying to change an element in an array:")
        arr[1]=34
    except ValueError as e:
        logging.error(f"raising {e}")
    
def main():
    arr=np.array([[12,23,34],[23,45,56],[13,33,67]])
    n_arr=read_array(arr)
    check_immutablity(arr)
    
    
if __name__=="__main__":
    main()
```
