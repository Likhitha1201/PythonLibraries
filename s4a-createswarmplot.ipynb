```python
""" 

    @Author: Likhitha S
    @Date: 25-10-2024 15:30
    @Last Modified by: Likhitha S
    @Last Modified time: 26-10-2024 15:30
    @Title:Write a program to draw a swarm plot of total bill against size  for a  given dataset 
    https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv 

"""


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def main():
    """

        Description: 
            This function is used to draw a graph using box plot.
        Parameters: 
            url is the input where we are going to get the dataset.
        return:
            It displays the box graph.

    """
    
    
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
    var= pd.read_csv(url)
    
    print(var)
    sns.swarmplot(x='size', y='total_bill', data=var)
    
    plt.title('Total Bill V/S Size')
    plt.xlabel('size')
    plt.ylabel('totalbill')
    
    plt.show()
    
if __name__=="__main__":
    main()
```
