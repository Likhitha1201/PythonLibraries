```python
""" 

    @Author: Likhitha S
    @Date: 25-10-2024 15:00
    @Last Modified by: Likhitha S
    @Last Modified time: 25-10-2024 15:00
    @Title:  Write a program to draw a violin plot of “species” against “sepal length” for a dataset 
    given in a url- https://github.com/mwaskom/seaborn-data/blob/master/iris.csv
   
     
"""


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def main():
    """

        Description: 
            This function is used to draw a graph using violin plot.
        Parameters: 
            url is the input where we are going to get the dataset.
        return:
            It displays the violin graph.

    """
    
    
    url = "https://github.com/mwaskom/seaborn-data/raw/master/iris.csv "
    iris_url = pd.read_csv(url)
    
    print(iris_url)
    plt.figure(figsize=(8,6))
    sns.violinplot(x='species', y='sepal_length', data=iris_url)
    
    plt.title('Violin Plot Of Sepal_length by Species')
    plt.xlabel('Species')
    plt.ylabel('Sepal Length(cm)')
    
    plt.show()
    
if __name__=="__main__":
    main()
```
